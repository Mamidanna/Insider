```python
#To check if the emails were sent outside organization and count the total numver of email receipients
def totEmails(emailIds):
    return len(emailIds.split(";"))


def checkOtherDomains(emailIds):
    totalEmailsCount = totEmails(emailIds)
    #print(totalEmailsCount)
    regexp = re.compile(r'@dtaa')
    matchedCount = len(re.findall(regexp,emailIds))
    #print(matchedCount)
    if totalEmailsCount > matchedCount:
        return True
    else:
        return False
```

```python
def preprocessUser(df,username):
    employeeDF = df
    dateFormat = "%m/%d/%Y %H:%M:%S"
    employeeDF["date"] = pd.to_datetime( employeeDF["date"],format=dateFormat)
    employeeDF['day'] = employeeDF['date'].dt.floor('D')


    most_common_pc = employeeDF\
        .groupby(['user', 'day', 'pc'])\
        .size()\
        .to_frame('count')\
        .reset_index()\
        .sort_values('count', ascending=False)\
        .drop_duplicates(subset=['user', 'day'])\
        .drop(columns=['count'])\
        .sort_values(['user', 'day'])\
        .groupby('user')\
        .pc\
        .agg(pd.Series.mode)\
        .rename('most_common_pc')

    employeeDF = employeeDF.merge(most_common_pc, left_on='user', right_on='user')
    employeeDF['email_out'] = employeeDF.to.apply(lambda x: "" if pd.isnull(x) else checkOtherDomains(x))
    employeeDF['tot_emails'] = employeeDF.to.apply(lambda x: 0 if pd.isnull(x) else totEmails(x))
    employeeDF['totcc_emails'] = employeeDF.cc.apply(lambda x: 0 if pd.isnull(x) else totEmails(x))
    employeeDF['totbcc_emails'] = employeeDF.bcc.apply(lambda x: 0 if pd.isnull(x) else totEmails(x))
    employeeDF['recepients'] = employeeDF['tot_emails'] + employeeDF['totcc_emails'] + employeeDF['totbcc_emails']

    employeeDF.drop(columns=['tot_emails'],axis=1,inplace=True) 
    employeeDF.drop(columns=['totcc_emails'],axis=1,inplace=True) 
    employeeDF.drop(columns=['totbcc_emails'],axis=1,inplace=True) 
    employeeDF

    employeeDF['is_usual_pc'] = (employeeDF['most_common_pc'] == employeeDF['pc'])

    is_work_time = (8 <= employeeDF.date.dt.hour) & (employeeDF.date.dt.hour < 17)
    early_hrs = (employeeDF.date.dt.hour < 8) 
    late_hrs =  (employeeDF.date.dt.hour >= 17) & (employeeDF.date.dt.minute > 0)
    file_has_exe = employeeDF.filename.str.contains(".exe", case=False)
    imp_file = (employeeDF.filename.str.contains(".doc", case=False)) | (employeeDF.filename.str.contains(".pdf", case=False)) | (employeeDF.filename.str.contains(".jpg", case=False)) |  (employeeDF.filename.str.contains(".zip", case=False))
    jobsites = (employeeDF.websitetype == 1)
    wikileaks = (employeeDF.websitetype == 2)
    keylogger = (employeeDF.websitetype == 3)
    #average_early
    #average_late
    #othersyystemsused
    #systemsusedoutsideofficehours
    #employeeDF['employee'][10] = "file.exe"
    #employeeDF['employee'][22] = "file.exe"
    #employeeDF['employee'][7224] = "file.exe"
    #employeeDF['employee'][7234] = "file.exe"

    employeeDF['is_work_time'] = is_work_time
    employeeDF['subtype'] = employeeDF['activity']
    employeeDF['early_hrs'] = early_hrs 
    employeeDF['late_hrs'] = late_hrs
    employeeDF['file_has_exe'] = file_has_exe
    employeeDF['imp_file'] = imp_file
    employeeDF['jobsites'] = jobsites
    employeeDF['wikileaks'] = wikileaks
    employeeDF['keylogger'] = keylogger
    length=len(employeeDF.index)

    employeeDF.to_csv(username+"-Before.csv")

    pc=[]
    work=[]
    user=[]
    ehrs=[]
    lhrs=[]
    daays=[]
    emailsout=[]
    attachments=[]
    recepients=[]
    usb=[]
    activity=[]
    file=[]
    imp_file=[]
    jobs=[]
    wiki=[]
    keylog=[]
    pc_count=0
    work_count=0
    act_count=0
    early_count=0
    late_count=0
    attach_count=0
    out_count=0
    emailsout_count=0
    totalemails_count=0
    recepients_count=0
    usb_count=0
    file_count=0
    imp_file_count=0
    jobs_count=0
    wiki_count=0
    keylog_count=0
    count=0
    daay=employeeDF['day'][0]
    #print(daay)
    employee=username
    firstEntry = True
    
    for i, j in employeeDF.iterrows():
        if j['day']==daay:            
            act_count+=1
            if j['is_usual_pc'] == False:
                pc_count+=1
            if j['is_work_time'] == False:            
                work_count+=1
            if j['activity'] == 'Connect' or j['activity'] == 'Disconnect':
                usb_count+=1                       
            if j['early_hrs'] == True:
                early_count+=j['early_hrs']
            if j['late_hrs'] == True:
                late_count+=j['late_hrs']
            if j['attachments'] > 0:
                attach_count+=j['attachments']
            if j['email_out'] == True:
                emailsout_count+=1
            if j['recepients'] > 0:
                recepients_count+=j['recepients']
            if j['file_has_exe'] == True:
                file_count+=1
            if j['imp_file'] == True:
                imp_file_count+=1
            if j['jobsites'] == True:
                jobs_count+=1
            if j['wikileaks'] == True:
                wiki_count+=1
            if j['keylogger'] == True:
                keylog_count+=1
            count+=1
        else:
            daays.append(employeeDF['day'][count-1])
            '''print(employeeDF['day'][count-1])
            print(str(count))'''
            user.append(employee)
            activity.append(act_count)
            pc.append(pc_count)
            work.append(work_count)            
            ehrs.append(early_count)
            lhrs.append(late_count)
            emailsout.append(emailsout_count)
            attachments.append(int(attach_count))        
            recepients.append(recepients_count)
            usb.append(usb_count)
            file.append(file_count)
            imp_file.append(imp_file_count)
            jobs.append(jobs_count)
            wiki.append(wiki_count)
            keylog.append(keylog_count)
            #print("Attach Count: "+str(int(attach_count))+"Emails Count: "+str(emailsout_count))
            '''act_count=0
            pc_count=0
            work_count=0
            early_count=0
            late_count=0        
            attach_count=0
            emailsout_count=0
            recepients_count=0
            usb_count=0
            file_count=0
            imp_file_count=0
            jobs_count=0
            wiki_count=0
            keylog_count=0'''
            daay=employeeDF['day'][count]            
            count+=1
            '''print(employeeDF['day'][count])
            print(str(count))
    
    print(str(count))
    print(employeeDF['day'][count-1])'''
    
    daays.append(employeeDF['day'][count-1])
    activity.append(act_count)
    pc.append(pc_count)
    user.append(username)
    work.append(work_count)
    ehrs.append(early_count)
    lhrs.append(late_count)
    attachments.append(attach_count)
    emailsout.append(emailsout_count)
    recepients.append(recepients_count)
    usb.append(usb_count)
    file.append(file_count)
    imp_file.append(imp_file_count)
    jobs.append(jobs_count)
    wiki.append(wiki_count)
    keylog.append(keylog_count)
    
    x=[daays,user,activity,pc,work,ehrs,lhrs,emailsout,attachments,recepients,usb,file,imp_file,jobs,wiki,keylog]
    x= [list(i) for i in zip(*x)]
    z = pd.DataFrame(x, columns=['day','user','activites','unusual_pc','unusual_work','early_hrs','late_hrs','emails_out','attachments','totalEmailRecepients','suspUSB','suspFileDownloads','impFileUpload','jobSites','hackSites','keyloggerSites'])
    #z.to_csv(employee+'.csv')
    #print(z)
```
